```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Dense,Flatten,Input,Reshape,Conv2D,MaxPooling2D

import numpy as np
import matplotlib.pyplot as plt

from tensorflow.keras.datasets import mnist
from tensorflow.keras.utils import to_categorical

from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.applications.resnet import ResNet50
```

```python
!unzip 4classimages.zip
```

```python
resnet_model = ResNet50(include_top=False, weights='imagenet', input_shape=(100,100,3), classes=4)
```

```python
resnet_model.summary()
```

```python
for layer in resnet_model.layers:  # for not train the parameters in resnet model
    layer.trainable = False
```

```python
resnet_model.summary()
```

```python
generator = ImageDataGenerator(rotation_range=10,shear_range=1.1)
```

```python
gen = generator.flow_from_directory("./data/data/",batch_size=200,target_size=(100,100))
```

```python
flat = Flatten()(resnet_model.output)
d1 = Dense(400,activation="tanh")(flat)
d2 = Dense(100,activation="tanh")(d1)
out = Dense(4,activation="softmax")(d2)
```

```python
model = Model(inputs = resnet_model.input, outputs = out)
```

```python
model.summary()
```

```python
model.compile(optimizer="adam",loss="categorical_crossentropy",metrics=["accuracy"])
```

```python
model.fit(gen,epochs=5,steps_per_epoch=20)
```
